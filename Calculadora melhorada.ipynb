```python
# Melhorias das funções para as operações matemáticas

def soma(n1, n2):
    return n1 + n2

def subtrai(n1, n2):
    return n1 - n2

def multiplica(n1, n2):
    return n1 * n2

def divide(n1, n2):
    if n2 != 0:
        return n1 / n2
    else:
        print("Erro: Divisão por zero!") #evitar que a divisão divida por 0 
        return None

# Função principal da calculadora
def calculadora():
    while True:
        print("+==================+")
        print("| MENU DE OPERAÇÕES |")
        print("|  1 - Somar        |")
        print("|  2 - Subtrair     |")
        print("|  3 - Multiplicar  |")
        print("|  4 - Dividir      |")
        print("|  0 - Sair         |")
        print("+==================+")

        op = int(input("Escolha uma operação: ")) #menu 
        
        if op == 0: #opção de encerrar calculadora quando escolhe a opção "O"
            print("Encerrando calculadora...")
            break
        elif op < 1 or op > 4: #verificar se a opção do menu escolhida é válida
            print("Opção inválida. Escolha novamente.") #se não for válida, exibe essa mensagem 
            continue
        
        n1 = float(input("Digite o primeiro número: "))
        n2 = float(input("Digite o segundo número: "))

        if op == 1: #estrutura mais detalhada para melhor experiência do usuário na hora de ver o resultado
            resultado = soma(n1, n2)
            print(f"O resultado da soma é: {resultado}")
        elif op == 2:
            resultado = subtrai(n1, n2)
            print(f"O resultado da subtração é: {resultado}")
        elif op == 3:
            resultado = multiplica(n1, n2)
            print(f"O resultado da multiplicação é: {resultado}")
        elif op == 4:
            resultado = divide(n1, n2)
            if resultado is not None:
                print(f"O resultado da divisão é: {resultado}")
        else:
            print("Opção inválida. Escolha novamente.")

if __name__ == "__main__":
    calculadora()
```
